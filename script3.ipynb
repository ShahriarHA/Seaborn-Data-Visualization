# Day 8
### Pair plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```

```python
sns.pairplot(df,vars=['total_bill','tip'],hue="sex")
```

```python
sns.pairplot(df,x_vars=['total_bill','tip'],hue='sex')
```

```python
sns.pairplot(df,y_vars=['total_bill','tip'],hue='sex')
```

```python
sns.pairplot(df,hue="sex",kind="reg",diag_kind="hist")# "scatter", "hist", "kde","reg"
```

# Day 9
### Heatmap

```python
plt.figure(figsize=[8,6])
sns.heatmap(df[['total_bill',"tip","size"]].corr(),annot=True,cmap="Blues")
plt.show()
```
