# Day 6 – Styling & Themes
### Goal: Make plots visually attractive.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
sns.set_style('whitegrid')
# sns.set_context('talk')
# sns.set_palette('coolwarm')
sns.scatterplot(x=df.total_bill,y=df.tip,style=df.sex,hue=df.sex)
plt.title('tip by total bill')
plt.show()
```

# Day 7 – Review & Mini Project
### Goal: Combine what I’ve learned.
##### Create a small "Restaurant Tips Dashboard" with:
###### 1. catter plot (Total Bill vs Tip)
###### 2. Histogram of Total Bills 
###### 3. Bar plot of Tips by Day 
###### 4. Use Seaborn themes and legends

```python
sns.set_style('darkgrid')
plt.title('tip by total bill')
sns.scatterplot(x=df.total_bill,y=df.tip,style=df.day,hue=df.sex)
plt.show()
```

```python
sns.set_style('dark')
plt.title('Histogram of total bill')
sns.histplot(df.total_bill)
plt.show()
```

```python
sns.set_style('dark')
plt.title('bar plot of tip by day')
sns.barplot(x=df.day,y=df.tip,errorbar=None)
plt.show()
```
