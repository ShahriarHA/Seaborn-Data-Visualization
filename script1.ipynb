# Day 2 - scatter and line plots

```python
sns.scatterplot(data=df,x='total_bill',y='tip')
```

```python
sns.scatterplot(x=df.total_bill,y=df.tip,hue=df.sex,style=df.smoker)
```

```python
sns.lineplot(x=df.total_bill,y=df.tip,hue=df.sex,style=df.smoker)
```

```python
sns.scatterplot(x=df.total_bill,y=df.tip,hue=df.day,style=df.sex)
```
