# Day 10 - FacetGrid & Multi-Plots
### Goal: Visualize multiple subsets.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# sns.get_dataset_names()
```

```python
df = sns.load_dataset("tips")
```

```python
df
```

```python
a = sns.FacetGrid(data=df,col="sex",row='smoker',hue="day")
a.map(sns.scatterplot,'total_bill','tip')
```

```python
b = sns.FacetGrid(data=df)
b.map(sns.scatterplot,"total_bill","tip")
```

```python
c=sns.FacetGrid(data=df,col="day",row="smoker",hue="sex")
c.map(sns.scatterplot,"total_bill","tip").add_legend()
```

# Day 12 â€“ Combine Seaborn + Matplotlib
### Goal: Mix the power of both.

```python
sns.boxplot(x='day', y='tip', data=df)
plt.title("Tip Distribution by Day")
plt.axhline(3.01, color='red', linestyle='--')
```
